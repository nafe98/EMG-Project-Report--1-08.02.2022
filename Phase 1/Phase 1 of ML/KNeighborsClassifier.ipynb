# Data Cleaning and Exploratory Data Analysis with New Data

```python
index_names = raw_data[ (raw_data['stimulus'] != raw_data['restimulus'])].index
raw_data.drop(index_names, inplace = True)
print(raw_data)
```

```python
index_names = raw_data[ (raw_data['repetition'] != raw_data['rerepetition'])].index
raw_data.drop(index_names, inplace = True)
print(raw_data)
```

```python
X = raw_data.iloc[:,0:10]
y = raw_data.stimulus
```

```python
# Splitting data for training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 39)
```

# KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
classifier11 = KNeighborsClassifier()
classifier11.fit(X_train, y_train)
```

```python
y_pred21 = classifier11.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
knn_accuracy11 = {accuracy_score(y_test, y_pred21)}
print(f"Accuracy is: {accuracy_score(y_test, y_pred21)* 100}")
```

```python
print('\033[1m' +'Classification report : '+ '\033[0m\n', classification_report(y_test, y_pred21))

cm = confusion_matrix(y_pred21, y_test)
print('\033[1m' +'Confusion Matrix : '+ '\033[0m')
sns.heatmap(cm, cmap = 'OrRd',annot = True, fmt='d')
plt.show()
```
